**MENGIRIM REQUEST KE API**

```python
import requests

url = "https://api.agify.io?name=annisa"
response = requests.get(url)

print(response)
print(response.text)
```

**MENGGUNAKAN DATA JSON**

```python
data = response.json()

print(data)
# {'name':'annisa','age':21}

print(data["name"])
#annisa

print(data["age"])
#21
```

**API QUOTES___KUTIPAN INSPIRATIF**

```python
import requests
url = "https://quotes.liupurnomo.com/api/quotes/random"
data = requests.get(url).json()
print(data["data"]["text"])
print("-", data["data"]["author"])
```

**LOOPING DENGAN API**

```python
import requests

for i in range(7):
  url = "https://quotes.liupurnomo.com/api/quotes?category=motivasi&page=1&limit=10"
  data = requests.get(url).json()

  print(f"--- Quote ke-{i+1} ---")
  print(data["data"][i+1]["text"])
  print(f"- {data["data"][i+1]['author']}")
  print()
```
