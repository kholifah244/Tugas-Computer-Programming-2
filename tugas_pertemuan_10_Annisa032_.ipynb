SOAL : Buat program Python di Google Colab yang menggunakan API Genderize.io untuk memprediksi gender berdasarkan nama yang diinputkan
pengguna.

```python
import requests

nama = input("Masukkan nama:")
url = f"https://api.genderize.io?name={nama}"
data=requests.get(url).json()

print("=== HASIL PREDIKSI GENDER ===")
print(f"Nama:{data['name']}")
print(f"Gender:{data['gender']}")
print(f"Prob:{data['probability']}")
print("============================")
```
