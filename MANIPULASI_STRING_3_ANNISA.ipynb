**FORMAT SPECIFIERS**

```python
nama = 'halo,saya %s npm %s'

print (nama % ('Annisa Nur Kholifah' , '2451012032'))
```

```python
template_1 = 'halo,saya {nama} npm {npm}'
template_2 = 'saya suka {} dan {}'

print (template_1. format (nama = 'Annisa Nur Kholifah' , npm = '2451012032'))
print (template_2. format ('makan' , 'buah'))

```

**STRING INTERPOLATION /F**

```python
nama = 'Annisa Nur Kholifah'
npm = '2451012032'

print (f'perkenalkan saya {nama} dan npm {npm}')
```
