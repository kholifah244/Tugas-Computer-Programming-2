**OPERASI IN PADA STRING**

```python
informasi = 'Bahwasanya besok libur sampai dua tahun kedepan'

print ('sublime text' in informasi)
print ('notepad++' in informasi)
```

**MEMOTONG STRING**

```python
nama = 'Annisa Nur Kholifah'

print (nama [-17])
print (nama [5])
print (nama [8])

```

**SLICING STRING**

```python
nama_npm = 'Annisa Nur Kholifah_2451012032'

print(nama_npm [0:6])
print(nama_npm [7:10])
print(nama_npm [11:19])
print(nama_npm [0:19])
print(nama_npm [20:])
```

**MENGHITUNG PANJANG STRING**

**MEMERIKSA KARAKTER AWAL STRING**

```python
no_npm_1 = '2111768975'
no_npm_2 = '2456827872'
no_npm_3 = '2589939037'

masuk_2025 = '25'

print (no_npm_1.startswith(masuk_2025))

print (no_npm_2.startswith(masuk_2025))

print (no_npm_3.startswith(masuk_2025))
```
