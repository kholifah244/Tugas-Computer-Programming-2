**STRING**

```python
nama = ' ANNISA NUR KHOLIFAH'
npm = '2451012032'
prodi = ' BISNIS DIGITAL'

print (nama)
print (npm)
```

**MANIPULASI STRING**

```python
print (nama,npm)
```

```python
print ('halo saya', nama, 'dengan npm' , npm)
```

```python
data_lengkap = nama + npm
print (data_lengkap)
```

```python
data_lengkap_banget = nama + npm + prodi
print (data_lengkap_banget)
```

**STRING + NON STRING**

```python
asal_kota = 'bandar lampung'
kode_pos = 35114

asal_kota_dengan_kode_pos = asal_kota + str (kode_pos)

print (asal_kota_dengan_kode_pos)
```

```python
asal_kota_x3 = asal_kota * 3
print (asal_kota_x3)
```
